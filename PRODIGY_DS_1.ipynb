### EDA:

```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt

csv_file = r'C:\Users\tessa\Downloads\Dataset1.csv'

# Read the CSV file into a Pandas DataFrame
df=pd.read_csv(csv_file)
df.dtypes
```

```python
print(df.info())
```

```python
#Descriptive Statistics
print(df.describe())
```

```python
# Separate the data into different series
districts = df['District']
states = df['State']
growth_in_population = df['Growth']
sex_ratio = df['Sex_Ratio']
literacy_rate = df['Literacy']
```

```python
df.head()
```

```python
df.tail()
```

### VISUALISATION:

```python

#bar chart for growth in population
plt.figure(figsize=(10, 6))
plt.bar(states, growth_in_population, color='skyblue')
plt.xlabel('States')
plt.ylabel('Growth in Population')
plt.title('Growth in Population Across States')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()
plt.show()

# histogram for sex ratio
plt.figure(figsize=(10, 6))
plt.hist(sex_ratio,bins=20, color='salmon', edgecolor='black')
plt.xlabel('Sex Ratio')
plt.ylabel('Frequency')
plt.title('Distribution of Sex Ratio Across Districts')
plt.tight_layout()
plt.show()

#bar chart for literacy rate
plt.figure(figsize=(10, 6))
plt.bar(states, literacy_rate, color='lightgreen')
plt.xlabel('States')
plt.ylabel('Literacy Rate')
plt.title('Literacy Rate Across States')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()
plt.show()
```
